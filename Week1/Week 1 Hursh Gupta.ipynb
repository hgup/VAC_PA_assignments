## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
fig = plt.figure(figsize = (12, 5))
uber_drivers['PURPOSE*']
distance_by_purpose = uber_drivers.groupby('PURPOSE*')['MILES*'].sum()

# creating the bar plot
plt.bar(distance_by_purpose.index , distance_by_purpose , color=COLOR_THEME)

plt.xlabel("PURPOSE")
plt.xticks(rotation=30, ha='right')

plt.ylabel("No. of MILES covered")
plt.title("Travelling Analysis")
plt.show()

# INFERENCE:
# Meeting and Customer Visits are the leading reasons for travelling
# whereas Airpor/Travel, Moving and Charity are the trailing reasons for travelling
```

## Q22. What percentage of Miles were clocked under Business Category and what percentage of Miles were clocked under Personal Category ? (3 points)

### Note:Use the original dataframe without dropping the 'NA' values.


```python
c=uber_drivers.groupby('CATEGORY*')['MILES*'].count().reset_index()
c['PERCENTAGE']=(c['MILES*'] / c['MILES*'].sum() * 100).map('{:.2f}%'.format)
c
```
